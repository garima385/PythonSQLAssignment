```python
country=coun_json[1]
```

```python
rows=[]
for y in range(len(country)):
    row={}
    #print(data1['name'])
    row['Capital']=country[y]['capitalCity']
    row['Id']=country[y]['id']
    row['IncomeId']=country[y]['incomeLevel']['id']
    row['incomeValue']=country[y]['incomeLevel']['value']
    row['IncomeISO']=country[y]['incomeLevel']['iso2code']
    row['Isco2Code']=country[y]['iso2Code']
    row['Latitude']=country[y]['latitude']
    row['LendingId']=country[y]['lendingType']['id']
    row['LendingValue']=country[y]['lendingType']['value']
    row['LendingISO']=country[y]['lendingType']['iso2code']
    row['RegionId']=country[y]['region']['id']
    row['RegionValue']=country[y]['region']['value']
    row['RegionISO']=country[y]['region']['iso2code']
    row['AdminRegionId']=country[y]['adminregion']['id']
    row['AdminRegionValue']=country[y]['adminregion']['value']
    row['AdminRegionISO']=country[y]['adminregion']['iso2code']
    row['Names']=country[y]['name']
    row['Longitude']=country[y]['longitude']
    rows.append(row)
```

```python
counF=pd.DataFrame(rows)
counF.head()
```

```python
import sqldf
```

```python
CF=sqldf.run('select * from counF join gdp1 on counF.Id=gdp1.Code')
CF.head()
```

```python
cols=[]
count=1
for column in CF.columns:
    if column=='index':
        cols.append(f'index_{count}')
        count+=1
        continue
    cols.append(column)
CF.columns=cols
```

```python
#List countries with income level of "Upper middle income"
UpIncome=sqldf.run('select CName as Country from CF where IncomeValue="Upper middle income"')
UpIncome.head(20)
```

```python
#List countries with income level of "Low income" per region.
LowIncome=sqldf.run('select CName as Country, RegionValue as Region from CF where IncomeValue="Low income" order by RegionValue, CName')
LowIncome.head(20)
```

```python
#Calculate cumulative/running value of GDP per region ordered by income from lowest to highest and country name.
CuGDP=sqldf.run('select CName as Country,RegionValue as Region, IncomeValue, round(cast(sum(COALESCE("2018", 0) + COALESCE("2019", 0) + COALESCE("2020", 0) + COALESCE("2021", 0) + COALESCE("2022", 0) + COALESCE("2023", 0)) as numeric),2) as CummulativeGDP from CF group by RegionValue order by CummulativeGDP asc, RegionValue asc')
CuGDP.head(20)
```

```python
#Calculate percentage difference in value of GDP year-on-year per country
CalGDP=sqldf.run('select CName as Country, round(CAST(100 * (("2019" - "2018")/ "2018") AS numeric), 2) AS "2018_2019",round(CAST(100 * (("2020" - "2019")/ "2019") AS numeric), 2) AS "2019_2020",round(CAST(100 * (("2021" - "2020")/ "2020") AS numeric), 2) AS "2020_2021",round(CAST(100 * (("2022" - "2021")/ "2021") AS numeric), 2) AS "2021_2022",round(CAST(100 * (("2023" - "2022")/ "2022") AS numeric), 2) AS "2022_2023" from CF')
CalGDP.head(20)
```

```python
Demo=sqldf.run('select *, (COALESCE("2018", 0) + COALESCE("2019", 0) + COALESCE("2020", 0) + COALESCE("2021", 0) + COALESCE("2022", 0) + COALESCE("2023", 0)) as Total from CF')
```

```python
cols=[]
count=10
for column in Demo.columns:
    if column=='index':
        cols.append(f'index_{count}')
        count+=1
        continue
    cols.append(column)
Demo.columns=cols
```

```python
#List 3 countries with lowest GDP per region
Demo1=sqldf.run('select *, ROW_NUMBER() OVER (PARTITION BY RegionValue ORDER BY Total ASC) as DemG from Demo')
LowGDP= sqldf.run('SELECT RegionValue as Region, CName as Country, ROUND(CAST(Total as numeric), 2) as GDPCumm from Demo1 where DemG <= 3 order by RegionValue, DemG')
LowGDP.head(20)
```
